```python
fileObject = open("input.txt", "r")
data = fileObject.read()
limst = [x for x in data if x != "\n" and x != " "]
```

```python
#A = X, A = Y, A = C
guide = []
for i in range(0,len(limst),2):
    guide.append([limst[i],limst[i+1]])
score = {}
#score["A"] = 1
score["X"] = 1
#score["B"] = 2
score["Y"] = 2
#score["C"] = 3
score["Z"] = 3
def RPS_Score(a,b):
    score = 0
    #where a is input of you, b is your enemy
    if a == "X": #rock chosen by me
        if b == "A":#rock
            score = 3
        elif b == "B": #paper
            score = 0
        else: #scissors
            score = 6
    elif a == "Y":# paper
        if b == "A":#rock
            score = 6
        elif b == "B": #paper
            score = 3
        else: #scissors
            score = 0
    elif a == "Z": # scissors
        if b == "A":#rock
            score = 0
        elif b == "B": #paper
            score = 6
        else: #scissors
            score = 3
    else:
        raise ValueError("Enemy cannot input item other than XYZ")
    return score
```

```python
#total score is the sum of your scores for each round. 
#The score for a single round is the score for the shape you selected 
#(1 for Rock, 2 for Paper, and 3 for Scissors) plus the score for the outcome of the round 
#(0 if you lost, 3 if the round was a draw, and 6 if you won)
total_score = 0
for i, j in enumerate(guide): # i is iterator, j is value
    total_score += score[j[1]] + RPS_Score(j[1],j[0])
```

```python
#A for Rock, B for Paper, and C for Scissors enemy
#X for Rock, Y for Paper, and Z for Scissors me
#X means lose, Y means draw, Z means win
#A = X, A = Y, A = C
def RPS_Outcome(a,b): #a is enemy, b is outcome
    final_score = 0
    #where a is input of you, b is your outcome
    if b == "X": #lose return score of just shape
        #find shape that loses to outcome A
        if a == "A":#rock
            final_score += score["Z"] #you choose scissors
        elif a == "B": #paper
            final_score += score["X"]
        else: #scissors
            final_score += score["Y"]
    elif b == "Y":# draw return score of shape plus 3
        final_score += 3
        if a == "A":#rock
            final_score += score["X"] #you choose paper
        elif a == "B": #paper
            final_score += score["Y"]
        else: #scissors
            final_score += score["Z"]
    elif b == "Z": # win return score of shape plus 6
        final_score += 6
        if a == "A":#rock
            final_score += score["Y"] #you choose paper
        elif a == "B": #paper
            final_score += score["Z"]
        else: #scissors
            final_score += score["X"]
    else:
        raise ValueError("Enemy cannot input item other than XYZ")
    return final_score
```

```python
total_score = 0
for i, j in enumerate(guide): # i is iterator, j is value
    total_score += RPS_Outcome(j[0],j[1])
```
